# Ajustes de curva

### Explicação

O ajuste de curva é uma técnica amplamente utilizada em diversas áreas da ciência, incluindo física, engenharia, biologia, economia e muitas outras. O objetivo do ajuste de curva é encontrar a curva que melhor se ajusta aos dados disponíveis, permitindo a previsão ou a compreensão do comportamento das variáveis estudadas.

Um dos métodos mais comuns para realizar o ajuste de curva é o método dos mínimos quadrados. Esse método é utilizado para encontrar a curva que minimiza a soma dos quadrados dos desvios entre os valores observados e os valores previstos pela curva.

Segundo o livro "Numerical Methods for Engineers and Scientists" de Amos Gilat e Vish Subramaniam, o método dos mínimos quadrados é utilizado em diversas aplicações, como no ajuste de curvas para análise de dados experimentais, no ajuste de modelos matemáticos para simulação de sistemas físicos e na identificação de parâmetros de sistemas dinâmicos.

Além do método dos mínimos quadrados, existem diversas outras técnicas utilizadas para o ajuste de curva. Por exemplo, o livro "Regression Analysis by Example" de Samprit Chatterjee, Ali S. Hadi e Bertram Price apresenta diversos métodos para a realização de regressão, incluindo regressão linear, regressão polinomial e regressão não linear.

Em resumo, o ajuste de curva é uma técnica importante utilizada em diversas áreas da ciência para encontrar uma equação que melhor descreva a relação entre duas ou mais variáveis. Existem diversas técnicas utilizadas para o ajuste de curva, e a escolha da técnica mais adequada depende do tipo de dados e da forma da curva.

## Formulação dos métodos dos mínimos quadrados

O método dos mínimos quadrados é uma técnica utilizada para encontrar a curva que melhor se ajusta aos dados disponíveis, **minimizando a soma dos quadrados dos desvios entre os valores observados e os valores previstos pela curva**. A fórmula do método dos mínimos quadrados para uma função linear é:

## $y = \alpha_1x + α_2 ≡ ax + b$


Onde:

**$y:$** é a variável dependente (ou resposta);   
**$x:$** é a variável independente (ou preditora);  
**$a:$** é o intercepto da reta, ou seja, o valor de y quando x é igual a zero;   
**$b:$** é a inclinação da reta, que indica a taxa de variação de y com relação a x.
Para encontrar os valores de a e b que melhor se ajustam aos dados, utiliza-se a seguinte fórmula para o cálculo dos valores de $a$ e $b$:   


## $ \alpha_1 = \frac{\sum(x * y) - \frac{\sum x * \sum y}{n}}{\sum xˆ2 - \frac{(\sum x)^2}{n}} $   


## **$\alpha_2 = \frac{\sum y - \alpha_1 * \sum x}{n}$**   


## $\phi(x) = \alpha_1x + \alpha_2 $   

Para funções não lineares, a fórmula do método dos mínimos quadrados é mais complexa e pode variar dependendo da forma da curva;   
Em geral, utiliza-se uma abordagem iterativa para encontrar os valores que minimizam a soma dos quadrados dos desvios entre os valores observados e os valores previstos pela curva.

# Exemplo de MMQ

## 1 - Implementação 1 - passo a passo

###1.1 - Obter o dataset (Conjunto de dados)

```python
# Criando um dataset

tabela = [[1,2,5,6],[4,8,9,10]]
print(tabela)
n = len(tabela[0])
print(f'O tamanho da tabela é {n}')
```

###1.2 - Destrinchar a fórmula em pequenas operações

Aplicando a primeira fórmula

### $ \alpha_1 = \frac{\sum(x * y) - \frac{\sum x * \sum y}{n}}{\sum xˆ2 - \frac{(\sum x)^2}{n}} $

Para isso, que separarmos a fórmula em pedaços. Por exemplo, iremos armazenar o valor da somatória em alguma lugar, vejamos

Aqui vamos fazer, na ordem, as seguintes operações:

$\sum x $  
$\sum y $  
$\sum x * y $   
$\sum x^2 $


###1.3 - Realizando as operações da fórmula

```python
# Pega o x e y da tabela
x = tabela[0]
y = tabela[1]

# Realiza as somatórias
soma_de_x = sum(x) # ∑x
print(soma_de_x) 
soma_de_y = sum(y) # ∑y
print(soma_de_y)
soma_de_x_vezes_y = sum([x[i] * y[i] for i in range(n)]) # ∑x*y
print(soma_de_x_vezes_y)
soma_de_x_ao_quadrado = sum([x[i]**2  for i in range(n)]) # ∑x^2
print(soma_de_x_ao_quadrado)
```

###1.4 - Realiza a seguinte operação $\frac{\sum x * \sum y}{n}$

```python
valor_soma_x_vezes_soma_y_por_n = (soma_de_x * soma_de_y)/n # (∑x∗∑y) / n
print(valor_soma_x_vezes_soma_y_por_n)
```


###1.5 - Realiza a seguinte operação $\frac{(\sum x)^2}{n}$

```python
soma_x_quadrado_por_n = soma_de_x**2/n
print(soma_x_quadrado_por_n)
```

###1.6 - Montando a consulta
Agora com os resultados, vamos construi nossa fórmula, simplesmente, substituindo as operações

###1.6.1 Montando o Alpha1

### $ \alpha_1 = \frac{\sum(x * y) - \frac{\sum x * \sum y}{n}}{\sum xˆ2 - \frac{(\sum x)^2}{n}} $

```python
alpha1 = (soma_de_x_vezes_y - valor_soma_x_vezes_soma_y_por_n) / (soma_de_x_ao_quadrado - soma_x_quadrado_por_n)
print(alpha1)
```

###1.6.2 - Montando Alpha2

## **$\alpha_2 = \frac{\sum y - \alpha_1 * \sum x}{n}$**   


```python
alpha2 = (soma_de_y - alpha1 * soma_de_x) / n
print(alpha2)
```

###1.7- Resultado

```python
print(f'{alpha1:.6f}x {alpha2:.6f}') # :.6f diz que a saída será formatada para 6 números depois da virgula
```

## 2 - Implementação 2 - Criando funções de soma

```python
# Criando um dataset

tabela = [[1,2,5,6],[4,8,9,10]]
print(tabela)
n = len(tabela[0])
print(f'O tamanho da tabela é {n}')
```

###2.1 - Implementação funções uteis em python

```python
def calcula_soma(colecao):
  if not isinstance(colecao, list) and not isinstance(colecao, tuple):
    print('Não é possível realizar a operação')
    return None
  soma = 0
  for item in colecao:
    soma += item
  return soma

def pegar_tamanho(colecao):
  if not isinstance(colecao, list) and not isinstance(colecao, tuple):
    print('Informe uma lista ou uma tupla')
    return None
    
  tamanho = 0
  for item in colecao:
    tamanho += 1
  return tamanho
```

###2.2 - Calculando cada operação de soma

```python
# Pega o x e y da tabela
x = tabela[0]
y = tabela[1]

# Realiza as somatórias
soma_de_x = calcula_soma(x) # ∑x
print(soma_de_x) 
soma_de_y = calcula_soma(y) # ∑y
print(soma_de_y)
soma_de_x_vezes_y = calcula_soma([x[i] * y[i] for i in range(n)]) # ∑x*y
print(soma_de_x_vezes_y)
soma_de_x_ao_quadrado = calcula_soma([x[i]**2  for i in range(n)]) # ∑x^2
print(soma_de_x_ao_quadrado)
valor_soma_x_vezes_soma_y_por_n = (soma_de_x * soma_de_y)/n # (∑x∗∑y) / n
print(valor_soma_x_vezes_soma_y_por_n)
soma_x_quadrado_por_n = soma_de_x**2/n
print(soma_x_quadrado_por_n)
```

###2.3 - Montando a fórmula

####Alpha 1

```python
alpha1 = (soma_de_x_vezes_y - valor_soma_x_vezes_soma_y_por_n) / (soma_de_x_ao_quadrado - soma_x_quadrado_por_n)
print(alpha1)
```

###Alpha 2

```python
alpha2 = (soma_de_y - alpha1 * soma_de_x) / n
print(alpha2)
```

###2.4 - Resultado

```python
print(f'{alpha1:.6f}x {alpha2:.6f}') # :.6f diz que a saída será formatada para 6 números depois da virgula
```

##3 - Implementação com função sum e Numpy

```python
import numpy as np

# Dados de exemplo
x = np.array([1, 2, 5, 6])
y = np.array([4, 8, 9, 10])

n = len(x) # número de pontos
sum_x = sum(x)
sum_y = sum(y)
sum_x_squared = sum([xi**2 for xi in x])
sum_xy = sum([x[i] * y[i] for i in range(n)])

# Encontrando os coeficientes a e b da linha de regressão
alpha1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
alpha2 = (sum_y - alpha1 * sum_x) / n

# Imprimindo os coeficientes a e b
print("Coeficiente angular (alpha1):", alpha1)
print("Coeficiente linear (alpha2):", alpha2)
```

##4 - Utilizando função em numpy

```python
import numpy as np

# Dados de exemplo
x = np.array([1, 2, 5, 6])
y = np.array([4, 8, 9, 10])

# Encontrando a equação da linha de regressão
alpha1, alpha2 = np.polyfit(x, y, 1)

# Imprimindo os coeficientes a e b
print("Coeficiente angular (a):", alpha1)
print("Coeficiente linear (b):", alpha2)
```

###Exemplo 2 - Com linha de tendência

```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.array([1, 2, 5, 6])
y = np.array([4, 8, 9, 10])

plt.title("Tempo de trabalho x Salário") 
plt.xlabel("x tempo de trabalho em anos") 
plt.ylabel("y salário vezes 1000") 
plt.plot(x,y, 'ob')

# Adicionando a linha de tendência
z = np.polyfit(x, y, 1) # Função polyfit realiza o cálculo de ajuste de curvas
p = np.poly1d(z)
plt.plot(x,p(x),"r--") # Adiciona a linha na figura criada

plt.show()
```

###Exemplo 3 - COm linha de tendência e seus valores

```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.array([1, 2, 5, 6])
y = np.array([4, 8, 9, 10])

plt.title("Tempo de trabalho x Salário") 
plt.xlabel("Tempo de trabalho em anos") 
plt.ylabel("Salário vezes 1000") 
plt.plot(x,y, 'ob')

# Adicionando a linha de tendência
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

# Adicionando o valor da linha de tendência
textstr = f'Linha de tendência: y = {z[0]:.6f}x + {z[1]:.6f}'
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
        verticalalignment='top')

plt.show()
```
