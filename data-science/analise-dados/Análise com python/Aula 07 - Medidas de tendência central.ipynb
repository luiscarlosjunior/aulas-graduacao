# Vamos fazer uma média simples
Até nesse momento do curso, aula 07, aprendemos algumas coisas importântes para a análise de dados, desde a introdução a estatística até as medida separatrizes. Aqui vamos começar com o cálculo simples da **média aritmética.**

Sem ajuda de nenhuma biblioteca, como poderiamos inserir esses dados e depois fazer a média?

```
# Cria uma lista comum em python
idades = [10, 15, 12, 13, 12, 15, 12]

# vamos exibir todos os nomes
for x in idades:
  print(x)
```

Com a lista criada de idades, teremos que fazer a média de uma forma 'manual'.


```
soma = 0

# Fazer um laço para somar todos as idades
for x in idades:
  soma += x

# Pegar a quantidade de itens
quantidade = len(idades)

# Fazer a média
print('A média é ', round(soma/quantidade, 2))
```



---

Por padrão, as operações que efetuamos em um **DataFrame** não o afetam. Isto é, sempre que renomeamos, alterando ou removemos uma coluna, linha, etc, o Pandas devolve um novo DataFrame sem aquele dado. Logo, o DataFrame original se mantém intacto. Para contornarmos isso o jeito mais fácil é atribuir aquele resultado a mesma variável.

```
# Vamos fazer a mesma coisa, mas passando o valor para o mesmo dataset, desta vez 
# sem usar o headp()
alunos = alunos.rename({'Nova Coluna':'Ultima Coluna'}, axis=1)
```

### Média

```
média_da_turma = alunos['média'].mean()
print(round(média_da_turma, 0))
```

```
# No exemplo acima foi mostrado um número com muitas casas decimais
# vamos arredondar para duas casas depois da virgula
média_total_dos_alunos = round(alunos['média'].mean(), 2)
```

Aqui, usamos a função round e passamos como parametro a média e depois quantas casas depois da virgula.

```
# Vamos exibir
média_total_dos_alunos
```

### Mediana

```
# Retorna a mediana da coluna média
alunos['Total Faltas'].median()
```

# Medidas separatrizes (Quartis)

Os quartis, como visto em aula, são medidas que distribui os valores em partes, aqui vamos usar o método quantile() do pandas, por padrão ele retorna o Quartil 2 (mediana), mas podemos passar qual quartil, decil ou percentil queremo por parametrô.

```
# Retorna o quartil 2 (0.5)
alunos['Total Faltas'].quantile()
```

```
# Retorna o quartil 1
alunos['Total Faltas'].quantile(0.25)
```

```
# Retorna o quartil 3
alunos['Total Faltas'].quantile(0.75)
```

Para os outros, é só passar o valor por paramêtro

```
# Retorna o decil 10
alunos['Total Faltas'].quantile(0.10)
```

```
# Retorn o percentil 95
alunos['Total Faltas'].quantile(0.95)
```
